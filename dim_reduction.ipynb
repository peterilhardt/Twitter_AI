```python
import numpy as np
import matplotlib.pyplot as plt
import pickle

from dimensionality_reduction import *
```

```python
with open('all_years_clean_sentiment2.pickle', 'rb') as f:
    all_years_clean = pickle.load(f)
    f.close()
with open('all_years_doc_term4.pickle', 'rb') as f:
    all_years_doc_term = pickle.load(f)
    f.close()
with open('all_years_doc_topic_df.pickle', 'rb') as f:
    all_years_doc_topic = pickle.load(f)
    f.close()
```

```python
# plot PCA scores of the 'all_years' tweets document-topic matrix
plot_PCA_2D(df = all_years_doc_topic, groups = all_years_clean['tweet_year'], 
            group_labels = all_years_clean['tweet_year'].unique(), random_state = 767, 
            file = 'pca_scores_all_years_doc_topic.png')
```

```python
# Scree plot for PCA of document-topic matrix
plt.figure(figsize = (10, 7))
pca2_var_explained = pca_perc_var_explained(all_years_doc_topic, n_components = 5, random_state = 767)
plt.plot(np.arange(1, 6), pca2_var_explained, color = 'darkred')
plt.xlabel('Number of Components', fontsize = 15)
plt.ylabel('Percent of Variance Explained', fontsize = 15)
plt.xticks(np.arange(1, 6), fontsize = 12)
plt.yticks(fontsize = 12)
plt.title('Scree Plot for Documents-Topics Matrix', fontsize = 18)
plt.savefig('pca_scree_all_years_doc_topic.png', dpi = 100, bbox_inches = 'tight')
plt.show()
plt.close()
```

```python
# plot PCA scores for 'all_years' tweets document-term matrix
plot_PCA_2D(df = all_years_doc_term, groups = all_years_clean['tweet_year'], 
            group_labels = all_years_clean['tweet_year'].unique(), random_state = 767)
```

```python
# plot t-SNE scores for document-topic matrix
plot_tsne(df = all_years_doc_topic, groups = all_years_clean['tweet_year'], 
          group_labels = all_years_clean['tweet_year'].unique(), random_state = 767, 
          file = 'tsne_all_years_doc_topic.png')
```
